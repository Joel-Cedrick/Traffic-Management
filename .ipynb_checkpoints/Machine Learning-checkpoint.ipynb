  """ Il s'agit d'un algorithme Supervisé  permettant de prevoir en sortie la vitesse et le debit des auto sur  sur une voie donnee
elle se base sur des donnees d apprentissage comportant : id_voie jour n_jour mois annee heure evenement meteo debit vitesse """
elle utilise les modules numpy,pandas, matplotlib, sckit learn ,
elle est accompagnée egalement du fichier donnees.csv
    

```python
import numpy as np
import pandas as pd 
from matplotlib import pyplot as plt
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import train_test_split
%matplotlib inline
```

```python
data=pd.read_csv( 'donnees.csv')

```

```python
df_x=data.iloc[:,1:-2]
df_y=data.iloc[:,-2:]
```

```python
x_train, x_test, y_train, y_test = train_test_split(df_x, df_y, test_size=0.3, random_state=4)
""" partage des donnees 70% pour l'entrainement et 30% pour le test """
```

```python

clf =  MLPRegressor(hidden_layer_sizes=(1000,),  activation='relu', solver='adam', alpha=0.001, batch_size='auto',
    learning_rate='constant', learning_rate_init=0.01, power_t=0.5, max_iter=10000, shuffle=True,
    random_state=0, tol=0.0001, verbose=False, warm_start=False, momentum=0.9, nesterovs_momentum=True,
    early_stopping=False, validation_fraction=0.1, beta_1=0.9, beta_2=0.999, epsilon=1e-08)
```

```python
clf.fit(x_train,y_train)
```

```python
pred=clf.predict(x_test)
```

```python
print(pred)
""" donnees predites """
```

```python
""" score du test """
clf.score(x_test,y_test )
```
